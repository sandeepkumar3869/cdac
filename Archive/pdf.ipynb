```python
k = dict()

for i in range(10):
    k[i+1] = random.randint(10,100)

colm = [i for i in k.keys()]

row = [i for i in k.values()]
row
colm



```

```python
import statsmodels
prob= []

for i in row:
    p1 = i/sum(row)
    prob.append(p1)

prob   


import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(8,6))

sns.barplot(x=colm,y=prob)
plt.title("probablity")

plt.xlabel('Product')
plt.ylabel('probablity of defective')

plt.show()

```
