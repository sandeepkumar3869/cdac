```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier,plot_tree

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/ml-basics/diabetes.csv')
data.head()
```

```python
data =data.drop(columns='PatientID')
```

```python
data.columns
```

```python
features = ['Pregnancies', 'PlasmaGlucose', 'DiastolicBloodPressure',
       'TricepsThickness', 'SerumInsulin', 'BMI', 'DiabetesPedigree', 'Age']
target = ['Diabetic']
```

```python
x = data[features].values
y = data[target].values
```

```python
model = DecisionTreeClassifier()
```

```python
xtrin,xtest, ytrain , ytest=train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
model.fit(xtrin,ytrain)
```

```python
tree.plot_tree(model)
```

```python
# Plot the tree
class_names=["yes","n0"]
plt.figure(figsize=(50, 50))  # Set figure size to make the tree larger
plot_tree(
    model, 
    filled=True, 
    rounded=True, 
    feature_names=features,  # Replace with your list of feature names
    class_names= class_names ,      # Replace with your list of class names
    proportion=True,
    fontsize=10
)
plt.show()
```

```python
ypred = model.predict(xtest)
```

```python
acc = accuracy_score(ytest,ypred)

acc
```

```python
pre = precision_score(ytest,ypred)
recall = recall_score(ytest,ypred)
matris = confusion_matrix(ytest,ypred)
```
