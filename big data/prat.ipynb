```python
from pyspark import SparkConf,SparkContext
```

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.master('local').appName("first").getOrCreate()
```

```python
df =spark.read.csv('2015-summary.csv', header=True)
```

```python
df.show()
```

```python
df.createOrReplaceTempView("ok")
```

```python
result = spark.sql("SELECT * from ok")
```

```python
result.show()
```

```python
result = spark.sql("SELECT DISTINCT(DEST_COUNTRY_NAME), sum(count) as sm FROM ok GROUP BY DEST_COUNTRY_NAME")
```

```python
result.show()
```

```python
from pyspark.sql.functions import col

# Filter rows where 'count' is even
filtered_df = df.filter(col('count') % 2 == 0)

# Show the filtered result
filtered_df.show()
```

```python
# Filter rows where 'count' is even and 'DEST_COUNTRY_NAME' is 'United States'
filtered_df = df.filter((col('count') % 2 == 0) & (col('DEST_COUNTRY_NAME') == 'Greece'))

# Show the filtered result
filtered_df.show()
```
