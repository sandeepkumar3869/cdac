Non-parametric ANOVA test

```python
#Read data from file
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
onestop_df = pd.read_csv('onestop.csv')
onestop_df.head(5)
```

```python
#See the distribution of the groups
import seaborn as sn
import matplotlib.pyplot as plt
sn.distplot(onestop_df['discount_0'], label = 'No Discount')
sn.distplot(onestop_df['discount_10'], label = '10% Discount')
sn.distplot(onestop_df['discount_20'], label = '20% Discount')
plt.legend();
```

```python
#Conducting the test
from scipy.stats import f_oneway

f_oneway( onestop_df['discount_0'],
          onestop_df['discount_10'],
          onestop_df['discount_20'])
```

```python
#P value is less than 5%
#We reject the null hypothesis
#That is mean sales value under different discount are different
```
